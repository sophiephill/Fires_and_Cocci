```python
#### Pool cases in meta:
cis = pd.read_csv("data/AttFullPeriod.csv")
raw = pd.read_csv("data/CasesPooledByPeriod.csv")

# Old way: I pooled observed/predicted cases using meta then used this approximation to derive the se of the %change
# From p.16 https://ai.stanford.edu/~ronnyk/2009controlledExperimentsOnTheWebSurvey.pdf
def getPercentChangeCiOLD(obs, obsSE, pred, predSE, upper = False):
    cvObs2 = (obsSE / obs)**2
    cvPred2 = (predSE / pred)**2
    pctDiff = 100 * (obs - pred) / pred
    sqrtBlock = 1.96 * np.sqrt(cvObs2 + cvPred2 - (1.96**2) * cvObs2 * cvPred2) 
    ci = (1 - sqrtBlock) / (1 - (1.96 * cvPred2) )
    if upper:
        ci = (1 + sqrtBlock) / (1 - (1.96 * cvPred2) )
    ci = (pctDiff + 1) * ci 
    return ci - 1

np.random.seed(1234)
def getPercentChangeCi(hexid, period, quantile):
    sub = raw[(raw.hexid == hexid) & (raw.period == period)].iloc[0]
    boot_preds = np.random.normal(sub["predicted"], sub["predSE"], 10000)
    boot_pchanges = 100 * (sub["observed"] - boot_preds) / boot_preds
    return np.quantile(boot_pchanges, quantile)

# Returns a formatted string with: X (95% CI: A, B) representing percent change + ci
# Sum predicted/observed in the period, use SE from meta to get ci for predicted per month
def percentChange(hexid, period):
    cases = raw[(raw.hexid == hexid) & (raw.period == period)].iloc[0]
    obs = cases["observed"]
    pred = cases["predicted"]
    sePred = cases["predSE"]
    seObs = cases["obsSE"]
    percChange = 100 * (obs - pred) / pred
    # Formula from census: https://www2.census.gov/programs-surveys/acs/tech_docs/accuracy/percchg.pdf
    # sePercChange = 100 * (obs/pred) * np.sqrt((varPred /(pred**2)) + (varObs /(obs**2)))
    #CI = [percChange - (1.96 * sePercChange), percChange + (1.96 * sePercChange)]
    CI = [getPercentChangeCi(hexid, period, 0.025), getPercentChangeCi(hexid, period, .975)]
    string = "{pchange:.1f} ({low:.1f}-{high:.1f})"
    string2= "obs: {o:.1f}, pred: {p:.1f}".format(o = obs, p = pred)
    return string.format(pchange = percChange, low = CI[0], high = CI[1]) + " / " + string2

def getCi(hexid, period):
    dat = cis[(cis.hexid == hexid )& (cis.period == period)]
    string = "{h} / {p}: {att:.1f} ({low:.1f}-{high:.1f}), Percent Change: "
    string = string.format(h = hexid, p = period, att = dat["ATT"].iloc[0], 
                           low = dat["CI.lower"].iloc[0], high = dat["CI.upper"].iloc[0])
    print(string + percentChange(hexid, period))

```

```python
#### Sum cases without pooling/weighting by variance 
raw = pd.read_csv("data/MonthlyAverageEffect.csv")
cis = pd.read_csv("data/AttFullPeriod.csv")
raw = raw[raw.hexid.isin(keepHexes) | (raw.hexid.str.contains("ALL"))]

# Returns a formatted string with: X (95% CI: A, B) representing percent change + ci
# Sum predicted/observed in the period, use SE from meta to get ci for predicted per month
def percentChange(hexid, period):
    startMonth = int(re.findall("(\\w+)-", period)[0]) 
    if startMonth == 0:
        startMonth += 1
    endMonth = int(re.findall("-(\\w+)", period)[0])
    cases = raw[(raw["month"] >= startMonth) & (raw["month"] <= endMonth)]
    
    predSe = cases.loc[cases.hexid == hexid, "CI.upper"].values - cases.loc[cases.hexid == hexid, "CI.lower"].values 
    predSe = predSe / (1.96 * 2)

    # First filter to the requested group. I sum the cases from the indiviudal data myself
    if "Over" in hexid: 
        cases = cases[~cases["hexid"].isin(under10counts)]
    elif "Under" in hexid:
        cases = cases[cases["hexid"].isin(under10counts)]
    elif "FR" in hexid: # Just considering one fire 
        cases = cases[cases.hexid == hexid]
    cases = cases[~cases.hexid.str.contains("ALL")]
    
    varPred = np.sum(predSe ** 2)
    varObs = np.var(cases["observed"])
    obs = sum(cases["observed"])
    pred = sum(cases["predicted"])
    percChange = 100 * (obs - pred) / pred
    # Formula from census: https://www2.census.gov/programs-surveys/acs/tech_docs/accuracy/percchg.pdf
    sePercChange = 100 * (obs/pred) * np.sqrt((varPred /(pred**2)) + (varObs /(obs**2)))
    
    CI = [percChange - (1.96 * sePercChange), percChange + (1.96 * sePercChange)]
    string = "{pchange:.1f} ({low:.1f}-{high:.1f})"
    string2= "obs: {o}, pred: {p:.1f}".format(o = obs, p = pred)
    return string.format(pchange = percChange, low = CI[0], high = CI[1]) + " " + string2

def getCi(hexid, period):
    dat = cis[(cis.hexid == hexid )& (cis.period == period)]
    string = "{h} / {p}: {att:.1f} ({low:.1f}-{high:.1f}), Percent Change: "
    string = string.format(h = hexid, p = period, att = dat["ATT"].iloc[0], 
                           low = dat["CI.lower"].iloc[0], high = dat["CI.upper"].iloc[0])
    print(string + percentChange(hexid, period))

getCi("ALL", "0-12")
getCi("ALL", "13-24")
getCi("ALL", "25-36")
getCi("ALL", "0-36")
```

```python
covars = pd.read_csv("../Synth_analysis/MonthlyCovData/MonthlyCovariatesWPlacebos20km.csv")
covars = pd.merge(covars, thexes20[["HEXID", "acres", "md_svrt"]], left_on = "hexid", right_on = "HEXID", how = "left")
attsOg = pd.read_csv("data/AttFullPeriod.csv")

def plotByVar(period, var, ax = None, pre = True, pointPlot = False, logTransform = True, ylabel = True): 
    covarNames = {"NDaysAbove5ms": "Wind (Number of Days Above 5m/s)", "tmax": "Max Temp (C)", 
                  'tmean': "Mean Temp (C)", "ppt": "Precipitation (mm)", 
                  "mesma": "Percent Vegetative Landcover", "population": "log Population", 
                  "CaseCount":"Pre-Fire Case Count", "acres": "Fire Burn Area (acres)"}
    covs = covars[covars["hexid"].isin(keepHexes) & covars["id"].str.contains("FR")]
    agg = np.mean
    if var == "CaseCount":
        agg = sum
        covPeriod = (covs["year"] < 0) 
    elif pre:
        covPeriod = (covs["year"] == -1) 
    else:
        covPeriod = (covs["year"] == 1) 
    covs = covs[covPeriod].groupby("hexid", as_index = False).agg(agg)[["hexid", var]]
    atts = attsOg[attsOg["hexid"].isin(keepHexes) & (attsOg.period == period)]
    atts = pd.merge(atts, covs[["hexid", var]], on = "hexid")
    atts = atts.sort_values(var)
    # Never log transform case count, always log transform population
    if (var == "population") | ((var != "CaseCount") and logTransform): 
        atts[var] = np.log(atts[var] + 0.5)
    gray = sns.color_palette("Greys")[5]
    lightgray = sns.color_palette("Greys")[4]
    blue = sns.color_palette().as_hex()[0]

    ylab = ""
    if ylabel:
        ylab = "Total Excess Cases (95% CI)"
    # Pointplots are misleading; they give equal x axis spacing which distorts the relationship. Don't use!
    if pointPlot:
        # need unique axis location
        dupes = atts[var][atts[var].duplicated()]
        k = 0.1
        for ind in dupes.index:
            atts.loc[ind, var] += k
            k += 0.1
        if ax is None: 
            plt.figure(figsize = (8, 6))
            ax = sns.pointplot(x = var, y = "ATT", hue = "hexid", data = atts, join = False, 
                               legend = False, palette = lightgray)
        else:
            sns.pointplot(x = var, y = "ATT", hue = "hexid", data = atts, join = False, 
                       legend = False, palette = lightgray, ax = ax)
        xs = ax.get_xticks()
        ax.legend().remove()
        ax.set_xticks([]);
    else:
        if ax is None:
            plt.figure(figsize = (8, 6))
            #ax = sns.scatterplot(x = var, y = "ATT", hue = "hexid", data = atts, legend = False, ax = ax, palette = pal)
#             ax = sns.regplot(x = var, y = "ATT", data = atts, ax = ax, color = lightgray, ci = None, 
#                              line_kws = {"color": blue})
            pal = ["gray" for h in atts["hexid"].unique()]
            ax = sns.scatterplot(x = var, y = "ATT", hue = "hexid", palette = pal,data = atts, ax = ax, legend = False)
            ax.set_ylabel(ylab, fontsize = 15)
            ax.set_xlabel(covarNames[var], fontsize = 15);
        else:
            sns.scatterplot(x = var, y = "ATT", hue = "hexid", data = atts, ax = ax, legend = False)
            ax.set_ylabel(ylab, fontsize = 30)
            if (var in ["acres", "population","CaseCount"]):
                xlab = covarNames[var] 
            elif pre or (var == "mesma"):
                xlab = "Pre-Fire " + covarNames[var]
            else:
                xlab = "Post-Fire " + covarNames[var]
            ax.set_xlabel(xlab, fontsize = 30);
        xs = atts[var]
    ax.tick_params(labelsize = 18)
    ebarPal = [blue if h in periodsigs else lightgray for h in atts["hexid"]]
    ax.errorbar(xs, atts["ATT"], yerr = getErrorbars(atts), ecolor = ebarPal, ls = "none", elinewidth = 4)
    ax.axhline(0, color = "black")
    ax.grid(color = 'black', linewidth = 0.5)
    #ax.set_xticklabels(round(atts[var], 1));
    #ax.set_title(covarNames[var]);
```
